```python
import fastf1
import pandas as pd

# Get the base directory path
fastf1.Cache.enable_cache('../data/raw')  # Cache in /data
```

```python
# Create an object with the names of the countries of the GPs per round
gp_countries = {
    1: 'Australia',
    2: 'China',
    3: 'Japan',
    4: 'Bahrain',
    5: 'Saudi_Arabia',
    6: 'Miami',
    7: 'Emilia_Romagna',
    8: 'Monaco',
    9: 'Spain',
    10: 'Canada',
    11: 'Austria',
    12: 'Great_Britain',
    13: 'Belgium',
    14: 'Hungary',
    15: 'Netherlands',
    16: 'Italy',
    17: 'Singapore',
    18: 'United States',
    19: 'Mexico',
    20: 'Brazil',
    21: 'Qatar',
    22: 'Abu_Dhabi'
}
```

```python
from fastf1.ergast import Ergast
ergast = Ergast()

standings = ergast.get_constructor_standings(season=2025, round=13)
```

```python
standings.content[0]
```

```python
# save the standings to a CSV file
standings_df = pd.DataFrame(standings.content[0])
standings_df.to_csv('../data/processed/standings/teams/constructor_standings_2025_Round_13_Belgium.csv', index=False)
```

```python
# Save all driver standings for the 2025 season from rounds 1 to 13 and save each to a CSV file
for round in range(1, 14):
    standings = ergast.get_driver_standings(season=2025, round=round)
    standings_df = pd.DataFrame(standings.content[0])
    standings_df.to_csv(f'../data/processed/standings/drivers/driver_standings_2025_Round_{round}_{gp_countries[round]}.csv', index=False)
```
